# Importing the Dependencies

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
# laoding the dataset to a pandas DataFrame
dataset = pd.read_csv(r"C:\Users\sydul\Downloads\boston_house_prices.csv")
```

```python
dataset.head()
```

```python
# checking the number of rows and columns in the dataframe
dataset.shape
```

```python
# check for missing values
dataset.isnull().sum()
```

```python
# statistical measures of the dataset
dataset.describe()
```

Understanding the correlation between features in the dataset

1. Positive Correlation

2. Negative Correlation

```python
correlation = dataset.corr()
```

```python
# constructing a heatmap to nderstand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Splitting the data and Target

```python
x=dataset.drop(['MEDV'],axis=1)
y=dataset['MEDV']
```

```python
print(x)
print(y)
```

Splitting the into Training data and Test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model x_train
model.fit(x_train,y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(x_train)
```

```python
print(training_data_prediction)
```

```python
# R swuared error
score_1 = metrics.r2_score(y_train,training_data_prediction)


# Mean Absolute error
score_2 = metrics.mean_absolute_error(y_train,training_data_prediction)
print('R squared error', score_1)
print('Mean Absolute error:',score_2)
```

visualizing the actual prices and predicted prices

```python
plt.scatter(y_train,training_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual prices vs Predicted prices')
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(x_test)
```

```python
# R swuared error
score_1 = metrics.r2_score(y_test,test_data_prediction)


# Mean Absolute error
score_2 = metrics.mean_absolute_error(y_test,test_data_prediction)
print('R squared error', score_1)
print('Mean Absolute error:',score_2)
```
